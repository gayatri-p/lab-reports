```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from tabulate import tabulate
import pandas as pd
```

```python
data1 = pd.read_excel('data.xlsx', sheet_name='mono')
ys = data1['f']
xs = data1['theta']
# plt.axhline(ys[8], color='k')
plt.plot(xs, ys, '-ko', label='Data points')
plt.xlabel('Phase difference (in degrees)')
plt.ylabel('Frequency (in kHz)')
plt.legend()
plt.tight_layout()
plt.savefig('images/mono.eps')
```

```python
data1 = pd.read_excel('data.xlsx', sheet_name='di')
ys = data1['f']
xs = data1['theta']
plt.plot(xs[:9], ys[:9], '-ro', label='Acoustic Branch')
plt.plot(xs[9:], ys[9:], '-bo', label='Optical Branch')
plt.axhline(ys[9], color='b', linestyle='-.', label='Min. Optical Frequency')
plt.axhline(ys[8], color='r', linestyle='--', label='Max. Acoustic Frequency')
plt.xlabel('Phase difference (in degrees)')
plt.ylabel('Frequency (in kHz)')
plt.text(15, 8, 'Acoustic Branch', color='r', rotation=37)
plt.text(90, 28, 'Optical Branch', color='b', rotation=45)
plt.text(63, 22.3, f'Band Gap = {ys[9]-ys[8]} kHz', color='k', fontsize=11)
plt.annotate(text='', xy=(60, ys[8]), xycoords='data', xytext=(60,ys[9]), textcoords='data', arrowprops=dict(arrowstyle='<->',connectionstyle="arc3", color='k', lw=2))
plt.legend()
plt.tight_layout()
plt.savefig('images/di.eps')
```
