```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
f1 = np.array([20.3, 25.5, 30, 41, 52, 68.5, 91, 114.5, 221, 365, 500, 730, 890])
f10e3 = np.array([1.3, 1.69, 2.545, 3.075, 3.6, 4.9, 5.75, 8.6, 30, 36, 58, 72, 106, 200.5, 278, 307.5, 400, 485, 565, 685, 960])*1000
f10e6 = np.array([1.2, 1.4, 1.6, 1.805, 2])*(10**6)

f = np.concatenate((f1, f10e3, f10e6))

vi = 0.2005
vo = np.array([0.764, 0.998, 1.16, 1.48, 1.72, 1.92, 2.08, 2.22, 2.5, 2.78, 3, 3.28, 3.42, 3.64, 3.72, 3.82, 3.88, 3.9, 3.94, 3.96, 3.96, 3.96, 3.96, 3.92, 3.88, 3.8, 3.44, 3.02, 2.86, 2.52, 2.24, 2.02, 1.72, 1.28, 1.04, 0.92, 0.8, 0.64, 0.57])

av = vo/vi
gain = 20*np.log10(av)
```

```python
plt.plot(f, gain, '-k.', label='Observed Gain')
plt.xscale('log')
plt.legend()

plt.gca().set_ylim(bottom=5)

cutoff = np.max(gain)/np.sqrt(2)
plt.axhline(y=cutoff, color='b', linestyle=':', lw=1)
x1, x2 = 48.4582, 712888
plt.plot([x1, x1], [0, cutoff], 'b:', linewidth=1)
plt.plot([x2, x2], [0, cutoff], 'b:', linewidth=1)

plt.text(900, 10, r"$\longleftarrow$Bandwidth$\longrightarrow$", fontsize=12)
plt.text(1700, 19, r"$20\,\log(A_V)_{max}\,/\,\sqrt{2}$", fontsize=10)
plt.xlabel('frequency in Hz (Logarithmic scale)')
plt.ylabel('Gain in dB: 20 log($A_V$)')

print('Mid frequency Gain:', np.max(gain))
print(f'Cut-off frequencies: {x1} Hz and {x2/1000} kHz')
print(f'Bandwidth: {(-x1+x2)/1000} kHz')
```

```python
headers = ['f', 'Vo', 'Gain (Av)', 'Gain (dB)']
print(tabulate(np.array([f, vo, av, np.round(gain,2)]).transpose(), headers=headers))
```
