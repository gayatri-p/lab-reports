```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
f1 = np.array([26, 31, 34, 36, 42.5, 61, 75, 87, 100, 125, 130.5, 140.5, 155, 175, 206, 300, 800])
f10e3 = np.array([2, 5, 7, 12, 15, 17, 20, 25, 30, 35, 50, 70 ,100, 130, 150, 170, 188, 233, 250, 280, 310, 345, 360, 415, 380, 430, 460, 530, 585, 650, 700, 840])*1000
f10e6 = np.array([1, 1.2, 1.4, 1.6, 1.8, 2])*(10**6)

f = np.concatenate((f1, f10e3, f10e6))

vi = 0.2
vo = np.array([0.8, 0.92, 0.98, 1.03, 1.14, 1.3, 1.36, 1.41, 1.46, 1.47, 1.48, 1.5, 1.52, 1.52, 1.52, 1.54, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.56, 1.54, 1.50, 1.48, 1.44, 1.42, 1.39, 1.36, 1.31, 1.3, 1.28, 1.2, 1.24, 1.17, 1.14, 1.08, 1, 0.97, 0.93, 0.82, 0.7, 0.596, 0.528, 0.472, 0.432, 0.392])

av = vo/vi
gain = 20*np.log10(av)
```

```python
# %matplotlib qt
%matplotlib inline
```

```python
plt.plot(f, gain, '-k.', label='Observed Gain')
plt.xscale('log')
plt.legend()

plt.gca().set_ylim(bottom=5)
plt.gca().set_xlim(left=10)

cutoff = np.max(gain)/np.sqrt(2)
plt.axhline(y=cutoff, color='b', linestyle='--', lw=1.5)
x1, y1, x2 = 784163, 12.63, 28.5827
plt.plot([x1, x1], [0, cutoff], 'b--', linewidth=1.5)
plt.plot([x2, x2], [0, cutoff], 'b--', linewidth=1.5)

plt.text(900, 10, r"$\longleftarrow$Bandwidth$\longrightarrow$", fontsize=12)
plt.xlabel('frequency in Hz (Logarithmic scale)')
plt.ylabel('Gain in dB ($V_o/V_i$)')

print('Mid frequency Gain:', np.max(gain))
print(f'Curtoff frequencies: {x1/1000} kHz and {x2} Hz')
print(f'Bandwidth: {(x1-x2)/1000} Hz')
```

```python
headers = ['f', 'Vo', 'Gain (Av)', 'Gain (dB)']
print(tabulate(np.array([f, vo, av, gain]).transpose(), headers=headers))
```
