```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
```

```python
def straight_line(x, m, c): return m*x+c
def fit_line(xs, ys):
    popt, pcov = curve_fit(f=straight_line, xdata=xs, ydata=ys)
    xFit = np.arange(min(xs),max(xs)+1e-4, 2e-4)
    m, c = popt[0], popt[1]
    return xFit, straight_line(xFit, m, c), (m,c), (np.diag(pcov)**0.5)
```

```python
def gaussian(x, A, mu, sigma, c): return A*np.exp(-(x-mu)**2/(2*sigma**2)) + c
def fit_gauss(xs, ys, p0=None):
    popt, pcov = curve_fit(f=gaussian, xdata=xs, ydata=ys, p0=p0, maxfev=5000)
    xFit = np.arange(min(xs),max(xs)+1e-4, 2e-4)
    A, mu, sigma, c = popt
    return xFit, gaussian(xFit, A, mu, sigma, c), (A, mu, sigma, c), (np.diag(pcov)**0.5)
```

# Mass absorption coefficient of Aluminium

```python
def exponential(x, A, m, c, d): return A*np.exp(-m*(x-c)) + d
def fit_exponential(xs, ys):
    popt, pcov = curve_fit(f=exponential, xdata=xs, ydata=ys)
    xFit = np.arange(min(xs)-0.1,max(xs)+0.8, 2e-4)
    A, m, c, d = popt
    print(pcov)
    return xFit, exponential(xFit, A, m, c, d), (A, m, c, d), (np.diag(pcov)**0.5)
```

```python
v1 = pd.read_excel('data/mca.xlsx', sheet_name='al')
fits = fit_exponential(v1['baseline'], v1['net'])
plt.figure(figsize=(6,5))
plt.plot(v1['baseline'], v1['net'], '-kx', label='Data points')
plt.plot(fits[0], fits[1], '--r', label='Exponential Fit')
plt.legend()
plt.ylabel('Net Count')
plt.xlabel('Thickness (cm)')
# plt.xlim(2.5,3.8)
# fwhm = 2.3548*fits[2][2]/fits[2][1]*100
# print(f'Resolution: {fwhm}')
label = f'Best-fit parameters for\n      '+  r'$y=A e^{-b(x-c)} + d$' + f'\nA = {fits[2][0]:.2f}' + r' $\pm$ ' + f'{fits[3][0]*1e-8:.2f}\n' \
        + r'b' + f' = {fits[2][1]:.2f}' + r' $\pm$ ' + f'{fits[3][1]:.3f}\n' \
        + r'c' + f' = {fits[2][2]:.2f}' + r' $\pm$ ' + f'{fits[3][2]*1e-6:.3f}\n' \
        + r'd' + f' = {fits[2][3]:.2f}' + r' $\pm$ ' + f'{fits[3][3]*1e-2:.2f}\n'
plt.text(0.14,0.1,label, fontsize=10,transform=plt.gcf().transFigure)
# res.append(fwhm)
plt.axhline(np.max(v1['net'])/2, color='k', linestyle=':')
plt.axvline(3.38, color='k', linestyle=':')
plt.savefig('images/al.eps')

# exponential(fits[2]
```
