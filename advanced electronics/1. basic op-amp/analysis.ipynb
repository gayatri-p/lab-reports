```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from tabulate import tabulate
```

# inverting amp gain

```python
inp1 = np.array([0.202, 0.411, 0.613, 0.809, 1.040, 1.09, 1.318, 1.656, 1.809, 2.007, 2.280])
out11 = -np.array([2.021, 4.12, 6.16, 8.10, 10.43, 10.69, 10.72, 10.72, 10.72, 10.72, 10.72])
out12 = -np.array([1.11, 2.302, 3.444, 4.55, 5.87, 6.15, 7.42, 9.32, 10.19, 11.31, 12.3])
gain11 = out11/inp1
gain11_avg = np.mean(gain11[:5])
gain12 = out12/inp1
gain12_avg = np.mean(gain12)
print(np.round(gain11,2))
print('Avg gain:', gain11_avg)
print(np.round(gain12, 2))
print('Avg gain:', gain12_avg)
```

```python
plt.figure(figsize =(6, 5))
plt.plot(inp1, out11, '-.b*', label='$A_v=-10.04$')
plt.plot(inp1, out12, '-rD', label='$A_v=-5.62$')
plt.axhline(y=0, color='k', linestyle='-', linewidth=1)
plt.axvline(x=0, color='k', linestyle='-', linewidth=1)
plt.xlabel(r'$V_{in}$ (V)')
plt.ylabel(r'$V_{out}$ (V)')
plt.legend(loc='upper right')
```

```python
def add_arrow(line, position=None, direction='right', size=20, color='k'):
    """
    add an arrow to a line.

    line:       Line2D object
    position:   x-position of the arrow. If None, mean of xdata is taken
    direction:  'left' or 'right'
    size:       size of the arrow in fontsize points
    color:      if None, line color is taken.
    """
    if color is None:
        color = line.get_color()

    xdata = line.get_xdata()
    ydata = line.get_ydata()

    if position is None:
        position = xdata.mean()
    # find closest index
    start_ind = np.argmin(np.absolute(xdata - position))
    if direction == 'right':
        end_ind = start_ind + 1
    else:
        end_ind = start_ind - 1

    line.axes.annotate('',
        xytext=(xdata[start_ind], ydata[start_ind]),
        xy=(xdata[end_ind], ydata[end_ind]),
        arrowprops=dict(arrowstyle="->", color=color),
        size=size
    )
```
