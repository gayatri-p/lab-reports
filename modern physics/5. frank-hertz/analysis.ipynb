```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
import pandas as pd
```

# dataset 1

```python
df1 = pd.read_csv('data/set1.csv')
ua1 = np.array(df1['V'])
ic1 = np.array(df1[' I'])
```

```python
plt.plot(ua1[:-1], ic1[:-1], '-k*')
a1, b1, c1 = 22, 40, 55
plt.axvline(x=a1, color='k', linestyle=':')
plt.axvline(x=b1, color='k', linestyle=':')
plt.axvline(x=c1, color='k', linestyle=':')
plt.xlabel(r'Accelerating Voltage $U_A$ (V)')
plt.ylabel(r'Collector Current $I_C$ (nA)')
print('Delta Us:', b1-a1, c1-b1, 'Avg:', (b1-a1+c1-b1)/2)
```

# dataset 3

```python
df3 = pd.read_csv('data/set3.csv')
ua3 = np.array(df3['V'])
ic3 = np.array(df3[' I'])
```

```python
plt.plot(ua3[:-10], ic3[:-10], '-k*')
a3, b3, c3 = 22.5, 40, 55
plt.axvline(x=a3, color='k', linestyle=':')
plt.axvline(x=b3, color='k', linestyle=':')
plt.axvline(x=c3, color='k', linestyle=':')
plt.xlabel(r'Accelerating Voltage $U_A$ (V)')
plt.ylabel(r'Collector Current $I_C$ (nA)')
print('Delta Us:', b3-a3, c3-b3, 'Avg:', (-a3+c3)/2)
```

```python
n = max(len(ua1), len(ua2), len(ua3))
ua1_ = np.concatenate((ua1, ['' for _ in range(n-len(ua1))]), axis=0)
ic1_ = np.concatenate((ic1, ['' for _ in range(n-len(ic1))]), axis=0)
ua2_ = np.concatenate((ua2, ['' for _ in range(n-len(ua2))]), axis=0)
ic2_ = np.concatenate((ic2, ['' for _ in range(n-len(ic2))]), axis=0)
ua3_ = np.concatenate((ua3, ['' for _ in range(n-len(ua3))]), axis=0)
ic3_ = np.concatenate((ic3, ['' for _ in range(n-len(ic3))]), axis=0)

table1 = np.array([ua1_, ic1_, ua2_, ic2_, ua3_, ic3_]).transpose()
print(tabulate(table1, tablefmt="latex"))
```
